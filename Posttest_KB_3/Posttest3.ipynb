# Visualisasi Dataset

## Mengimport Library

- **'import matplotlib.pyplot as plt'**, digunakan untuk memvisualisasikan data dalam bentuk Bar Chart dan Pie Chart
- **'import pandas as pd'**, digunakan untuk membaca dataset csv
- **'import seaborn as sns'**, digunakan untuk memvisualisasikan dataset dalam bentuk Heatmap dan Histogram

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

## Membaca file csv menggunakan encoding 'latin-1'

Encoding default membaca file csv yang digunakan pandas adalah **'UTF-8'**, mengubah encoding pandas menjadi **'latin-1'** agar file bisa dibaca oleh pandas

```python
df = pd.read_csv("BELGIUM_POPULATION_STRUCTURE_2018.csv", encoding='latin-1')
```

## Membuat Variabel Nationality

Membuat variabel **'Nationality'** untuk menampung semua data yang ada pada kolom *'NATIONALITY'*

```python
Nationality = df["NATIONALITY"].value_counts()
```

## Pie Chart

Memvisualisasikan data Nationality menggunakan Pie Chart untuk melihat seberapa banyak populasi penduduk Belgia yang berkewarganegaraan Belgia(Belgian atau Alien), Menggunakan *'labels=Nationality.index'* untuk memberikan pada label untuk mengetahui suatu bagian itu milik siapa, kemudian *'startangle=90'* digunakan agar titik mulai pembagian/pemotongan data dimulai dari 90 derajat, *autopct="%.2f%%"* digunakan untuk memformat persentase menjadi 2 angka dibelakang koma, *explode=(0, 0.1)* digunakan untuk memisahkan *'Alien'* sebesar 10% dari jari-jari lingkaran, dan *'wedgeprops={'linewidth': 2, 'edgecolor': 'black'}'* digunakan untuk memberikan ketebalan outline sebesar 2 dan warna outline menjadi hitam

```python
plt.pie(Nationality, labels=Nationality.index, startangle=90, autopct="%.2f%%", explode=(0, 0.1), wedgeprops={'linewidth': 2, 'edgecolor': 'black'})
plt.title("Persentase Kewarganegaraan Penduduk Belgia")
plt.show()
```

Dari data diatas dapat dilihat bahwa penduduk Belgia yang berkewarganegaraan **'Belgian'** sebesar 66.55% sedangkan **'Alien'** atau bukan berkewarganegaraan Belgia sebesar 33.45%, dan **'Alien'** dikeluarkan dari piechart untuk memisahkan data

## Heatmap

Membuat variabel **'belgium_population_corr'** untuk menampung seluruh korelasi antar kolom yang bertipe numerik dan ditampilkan dalam bentuk Heatmap, mengatur skema warna yang akan digunakan dalam heatmap, dalam kasus ini, warna yang digunakan adalah **'magma'**, kemudian mengaktifkan penambahan angka-angka di dalam sel heatmap untuk menunjukkan nilai korelasi yang  menggunakan *'annot=True'*.

```python
belgium_population_corr = df.corr()
sns.heatmap(belgium_population_corr, cmap="magma", annot=True)
plt.title("Hubungan/Korelasi Antar Kolom Bertipe Numerik")
plt.show()
```

Dari data diatas dapat dilihat bahwa kolom yang memiliki korelasi yang kuat atau korelasi positif memiliki warna terang, seperti *'MUNICIPALITY REFNIS CODE'* dengan *'DISTRICT REFNIS CODE'* dan *'PROVINCE REFNIS CODE'*, begitu pula pada kolom *'DISTRICT REFNIS CODE'* memiliki korelasi yang kuat dengan *'MUNICIPALITY REFNIS CODE'* dan *'PROVINCE REFNIS CODE'*, lalu *'PROVINCE REFNIS CODE'* memiliki korelasi yang kuat dengan *'DISTRICT REFNIS CODE'* dan *'PROVINCE REFNIS CODE'* sebesar **'1'**, kemudian untuk kolom yang memiliki korelasi paling lemah atau korelasi negatif ada pada *'AGE'* dan *'POPULATION'* sebesar **-0.19**

## Bar Chart

Menampilkan jumlah penduduk dari setiap wilayah menggunakan Bar Chart untuk melihat perbedaan jumlah penduduk, menggunakan **'plt.xticks(rotation = 0)'** agar keterangan yang berada di sumbu x tidak berotasi karena sebelumnya berotasi 90 derajat, dan menggunakan warna merah sebagai penanda untuk jumlah penduduk terbanyak, kemudian diikuti kuning dan jumlah penduduk terendah menggunakan warna hijau

```python
plt.xticks(rotation = 0)
plt.bar(df["REGION NAME"].value_counts().index, df["REGION NAME"].value_counts(), color=["red", "yellow", "green"])
plt.title("Jumlah Penduduk dari Setiap Wilayah")
plt.xlabel("Nama Region")
plt.ylabel("Jumlah Penduduk")
plt.show()
```

Dari data diatas dapat dilihat bahwa **'Flemish Region'** memiliki jumlah penduduk terbanyak sekitar 240000 dengan diikuti **'Walloon Region'** dengan jumlah sekitar 190000 dan dengan jumlah penduduk yang paling sedikit ada pada **'Brussels-Capital Region'** dengan jumlah sekitar 250000

## Histogram

Menampilkan jumlah masyarakat pada rentang umur tertentu menggunakan histogram

```python
sns.histplot(df["AGE"])
plt.xlabel("Umur")
plt.ylabel("Jumlah Masyarakat")
plt.title("Jumlah Masyarakat pada rentang umur tertentu")
plt.show()
```

Dari data diatas dapat dilihat pada rentan umur sekitar 0 hingga 17 tahun memiliki jumlah yang hampir sama, kemudian mengalami kenaikan jumlah masyarakat pada umur sekitar 19 sampai 39 tahun, lalu jumlah masyarakat pada umur sekitar 45 sampai 60 tahun merupakan jumlah yang terbanyak dan jumlah yang paling sedikit ada pada masyarakat berumur sekitar 95 sampai 98 tahun
